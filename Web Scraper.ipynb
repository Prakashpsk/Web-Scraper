```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
import csv
```

```python
url = 'https://www.amazon.com/s?k=phone&crid=E4KO8UEKG73D&sprefix=phone%2Caps%2C1087&ref=nb_sb_noss_1'
```

```python
headers = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36 Edg/122.0.0.0',
    'accept-language': 'en-US,en;q=0.5'
}
```

```python
webpage = requests.get(url, headers=headers)
```

```python
soup=BeautifulSoup(webpage.content,'html.parser')
```

```python
soup.find('span', attrs={'class':"a-size-base s-underline-text"}).text
```

```python
# Open CSV file for writing
with open('amazon_products1.csv', 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Title', 'Price', 'Rating'])

    # Find all products on the page
    products = soup.find_all('div', class_='s-result-item')

    # Loop through each product
    for product in products:
        try:
            title = product.find('span', class_='a-size-medium').text.strip()
        except AttributeError:
            title = 'N/A'

        # Not all products have prices or ratings, so we need to handle None values
        price_element = product.find('span', class_='a-offscreen')
        price = price_element.text.strip() if price_element else 'N/A'

        rating_element = product.find('span', class_='a-size-base s-underline-text')
        rating = rating_element.text.strip() if rating_element else 'N/A'

        # Write data to CSV file
        writer.writerow([title, price, rating])

print("CSV file created successfully!")
```
